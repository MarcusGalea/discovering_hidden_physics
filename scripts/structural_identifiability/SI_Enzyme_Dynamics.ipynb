```julia
using Revise, DifferentialEquations, ModelingToolkit, StructuralIdentifiability, Catalyst
```

```julia
#Create micaelis model(without hydrolysis)
rn = @reaction_network begin
    @parameters w_E w_S w_ES
    ka, S + E --> ES
    kd, ES --> E + S
end
odesys = convert(ODESystem, rn)
```

```julia

@unpack E,S,ES,w_E, w_S, w_ES, ka, kd= rn
measured_quantities_1 = [E, S, ES] #normal output
measured_quantities_2 = [w_E*E + w_ES*ES + w_S*S] #Octet output
```

```julia
res = assess_identifiability(odesys, measured_quantities = measured_quantities_1)
```

```julia
res = assess_identifiability(odesys, measured_quantities = measured_quantities_2, known_ic = [E, S, ES])
```

```julia
find_identifiable_functions(odesys, measured_quantities = measured_quantities_2, known_ic = [E, S, ES])
```

```julia
res = assess_identifiability(odesys, measured_quantities = measured_quantities_2, known_ic = [E, S, ES], funcs_to_check  = [ka/ kd])
```

# Hydrolysis

```julia
rn_hydrolysis = @reaction_network begin
    @parameters w_E w_S w_ES w_P
    ka, S + E --> ES
    kd, ES --> E + S
    kc, ES --> E + P
end
odesys_hydrolysis = convert(ODESystem, rn_hydrolysis)
@unpack P, w_P, kc = rn_hydrolysis
measured_quantities_3 = [E, S, ES, P] #normal output
measured_quantities_4 = [w_E*E + w_ES*ES + w_S*S + w_P*P] #Octet output
```

```julia
res_hydro = assess_identifiability(odesys_hydrolysis, measured_quantities = measured_quantities_3, known_ic = [E, S, ES, P])
```

```julia
res_hydro = assess_identifiability(odesys_hydrolysis, measured_quantities = measured_quantities_4, known_ic = [E, S, ES, P])
```
