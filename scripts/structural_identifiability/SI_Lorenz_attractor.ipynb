```julia
using Revise, DifferentialEquations, ModelingToolkit, StructuralIdentifiability
```

```julia
#Lorenz attractor
@independent_variables t
@parameters σ, ρ, β
@variables x(t), y(t), z(t)
D = Differential(t)
eqs = [D(x) ~ σ * (y - x),
       D(y) ~ x * (ρ - z) - y,
       D(z) ~ x * y - β * z]
@named sys_known = ODESystem(eqs, t, [x, y, z], [σ, ρ, β])
measured_vars = [x, y]
sys_known = complete(sys_known)
```

```julia
res = assess_identifiability(sys_known, measured_quantities = measured_vars)
```
